# EDA

## Data Cleaning

Datetime is already set and parsed as an index, we will remove non available values, duplicates and set an hourly frequency


```python
# Checking for NaN values

pjm_data = pjm_data.rename(columns={"PJMW_MW":"MW"})

num_nan_values = pjm_data["MW"].isna().sum()
print("NaN Values:",num_nan_values)
```

```python
pjm_data = pjm_data.loc[~pjm_data.index.duplicated()]

print(pjm_data.shape)
```

```python
pjm_data = pjm_data.asfreq("H")
# back fill na values na value
print(pjm_data.MW.isna().sum())
pjm_data.MW = pjm_data.MW.fillna(method="bfill")
print(pjm_data.MW.isna().sum())
```

Cool ! All values are available, let's now take a look at the data stats

## Plotting Weekly, Monthly and Yearly Data

```python

# Create day,week, month and year feature to use later in forecasting
pjm_data["day"] = pjm_data.index.dayofweek
pjm_data["week"] = pjm_data.index.weekofyear
pjm_data["month"] = pjm_data.index.month
pjm_data["year"] = pjm_data.index.year
pjm_data["dayofyear"] = pjm_data.index.dayofyear
pjm_data["hour"] = pjm_data.index.hour

print(pjm_data.head())


pjm_week = pjm_data.MW.resample('W').sum()
pjm_month = pjm_data.MW.resample("M").sum()
pjm_year = pjm_data.MW.resample("Y").sum()
pjm_day = pjm_data.MW.resample("D").sum()
print(pjm_year.head())
```

# Stationarity & Seasonality

We will test our data against stationarity and seasonality

## Stationarity:

Stationarity means that the timeseries can't be represented through a unit root, in other words, the means and covariance are time dependent.
We will first take a quick look at the rolling mean and the standard deviation of the last 500 observations, then we will use the **Augmented Dickey-Fuller** test with the null hypothesis $H_0$ being that the data is non stationary.

```python
# Rolling mean and std

rolling_mean = pjm_data.tail(500).MW.rolling(12).mean()
rolling_std = pjm_data.tail(500).MW.rolling(12).std()


plt.figure(figsize=(10,10))
plt.plot(pjm_data.tail(500).MW)
plt.plot(rolling_mean,color="red")
plt.plot(rolling_std,color="green")
plt.title("Rolling Mean and Rolling STD")
plt.show()
```

We can see that through the moving average the data seems non stationary, however this is not a good test, let's look at the results from the AFD test

```python
#Augmented Dickey Fuller Test
print(pjm_data.MW.isna().sum())
afd_results = sts.adfuller(pjm_data.MW)

print("AFD p-value:",afd_results[1])
```

We can see that indeed our data is stationary because $p<0.05$ for $95$% confidance interval, we don't need further transformations

## Seasonality and STL Decomposition

We can represent a timeseries in two ways: 

An additive model:
    $$ x(t) = Tr(t) + S(t) + \epsilon(t)    $$

or as a multiplicative model: 
$$  x(t) = Tr(t) * S(t) * \epsilon(t) $$

Where $x(t)$ is our time series, $Tr(t)$ is the trend, this can be for example a quadratic trend or an upward linear trend or really any kind of trend that tells us how the data is actually moving.

$S(t)$ is the seasonal component which tells if our data has some patterns linked to given period of time $T$.

$\epsilon(t)$ represents the residuals (the error term).

We will use the STL (Seasonal and Trend decomposition using Loess), **Loess** is a smoothing algorithm, basically it combines a linear regression along with a local regression, thus we can have a very flexible regression for the trend. The STL technique then use this trend created by Loess and substracts it from the time series, we get a detrended series which is the seasonal component + trend. Fortunately this is easy to do with statsmodel 

```python
from stldecompose import decompose
stl = decompose(pjm_data.MW, period = 24*365)
stl_plt = stl.plot()
```

Indeed the the trend is somehow linear and all seasonal components were successfully captured

# PCA Decomposition

```python
# PCA
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
```

```python
principalComponents = pca.fit_transform(pjm_data)
pca_df = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
# Getting info and plotting

print(pca_df.head())

plt.figure(figsize=(15,15))
plt.scatter(pca_df.iloc[:,0],pca_df.iloc[:,1])
plt.title("PCA Decomposition")

plt.show()
print("Explained Variance:",pca.explained_variance_ratio_)
```

As we can see, a high percentaage of the variance is explained using only 2 components.
